```python
from tensorflow import keras
from typing import List
from tensorflow.keras.preprocessing.text import Tokenizer  # Đường dẫn import đã sửa

sentence = ["John likes to watch movies. Mary likes movies too."]

def print_bow(sentence: List[str]) -> None:
    # chỉ cần 2 câu lệnh sau đã có thể xây dựng đc BoW
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(sentence)

    # Chuyển đổi mỗi câu thành một chuỗi số nguyên dựa trên từ điển đã học ở trên
    sequences = tokenizer.texts_to_sequences(sentence)
    print('in ra sequences:\n',sequences)
    print()

    # Trả ra từ điển các từ và chỉ số tương ứng. Đây là một Python dictionary.
    word_index = tokenizer.word_index
    print('in ra word_index:\n',word_index)
    print()

    bow = {}

    for key in word_index:
        bow[key] = sequences[0].count(word_index[key])

    print(f"Bag of word sentence 1:\n{bow}")
    print()
    print(f"We found {len(word_index)} unique tokens.")

print_bow(sentence)
```

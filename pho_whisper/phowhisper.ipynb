```python
import soundfile as sf
import IPython.display as ipd

from audio_preprocess import preprocess_audio
from asr_inference import load_model, transcribe_audio
from text_postprocess import postprocess_text
```

```python
file_path = "/content/audio.wav"
audio = preprocess_audio(file_path)
sf.write("cleaned.wav", audio, 16000)
ipd.display(ipd.Audio("cleaned.wav", rate=16000))

model, processor = load_model()
text = transcribe_audio(audio, model, processor)
text = postprocess_text(text)
# Kết quả
print(text)
```
